```python
"""The Lion King text game by Carmen Neier
Date: 10/24/2021
No bugs have been identified.

This text game is based on the famous Disney movie, The Lion King. It takes the player
through three difference scenes, starting in the Pride Lands, moving on to the Hakuna Matata
dessert, and finally returning to the Pride Lands to win
back the kingdom from the hyenas and evil uncle. In each scene the player faces
decisions which either bring them further in successfully completing the game or
result in a game over. No prior knowledge of the movie is needed to play the game;
however it is beneficial to appreciate the story line.  

"""

#importing extension time
import time

#defining function scene 1 
def scene_1():
    
    #introducing the game and prompting the player to start
    print(f"""{('*^')*40} 

    Hello and welcome! Embark on a journey to the Pride Lands of
    Africa and play your way through one of Disney's most famous stories. 

{('*^')*40} """)

    #delayed print for better flow
    time.sleep(3)
    
    #prompting user to continue game
    input("< Press ENTER to continue. >\n")
    
    #prompting user to pick a character name
    user_name = input(prompt = f"""
Your character in this game is a lion. Enter your character name to start.\n""").capitalize()

    #scene 1 introduction and prompting for first decision 
    print(f"""
You are a lion cub called {user_name}, and your father, Mufasa is the king of the savanna. 
You spend your days chasing other animals around the kingdom with your best 
friend Nala. One day your tremendous roar (which you have been practicing skillfully) 
causes a massive stampede of wildebeest to come down the mountain heading directly 
towards you. 

What do you do?""")

    #delayed user prompt for better story flow
    time.sleep(5)
    
    #user prompted to decide how to proceed in game
    decision_stampede = input(prompt = """
        1 climb a tree hoping the stampede will pass underneath you
        2 stand your ground and give them your best roar 
    
    """)

    ################################################################################################
    #developing a conditional statement where user either
    #progresses to next scene or
    #results in game over
    ################################################################################################

    if decision_stampede == '1' or "climb" in decision_stampede or "tree" in decision_stampede:
        print(f"""
Great, {user_name}! You managed to jump up and hold onto a branch in time for the stampede to
pass below you. However your father who is trying to rescue you, is killed in the process by 
your evil uncle. Your uncle makes you believe you are at fault.
        
'Run away {user_name}', he sais, 'run away, and never come back...'
        """)
     
        input("< Press ENTER to continue. >\n")
        
        scene_2()
        
    elif decision_stampede == '2' or "roar" in decision_stampede or "stand" in decision_stampede:
        print("\n You feel the stampede roll over you and you black out.")
        
        fail()
            
    else: 
        print("Invalid input. Please try again.")
        
        scene_1()

#defining function for scene 2
def scene_2():
    
    #print introduction to scene 2
    print("You start a new life far away from the Pride Lands with two new friends Timon and Pumba.\n")
    
    #delayed print for better flow
    time.sleep(2)
    
    #print statement to set the stage for decision point in scene 2
    print("Hakuna Matata! You are a vegetarian now. You inquire a selection of colored bugs to eat...\n")
    
    #list of bugs, [color, poison index] 1 = poisonous, 0 = not poisonous
    bugs = [['blue', 1], ['red', 0], ['purple', 1], ['orange', 0], ['black', 0]]

    #printing list of bug colors for user to chose from
    for color, poison_index in bugs:
        print(color)

    time.sleep(2)

    #user prompted to chose a bug to eat
    bug = input(prompt = 'Which bug do you chose?\n')

    #conditional statements to determine the outcome of the bug chosen and either
    #send user to scene 3 or fail
    if bug.lower() == 'blue' or bug.lower() == 'purple':
        print('Your stomach starts burning, this bug was lethal for young lion cubs.')
        fail()
        
    elif bug.lower() == 'red' or bug.lower() == 'orange' or bug.lower() == 'black':
        print('Mhh.. slimey yet somewhat satisfying.\n')
        scene_3()
    
    else:
        print('Invalid entry. Try again')
        scene_2()
        
#defining function for scene 3
def scene_3():
    
    ################################################################################################
    #while loop to demonstrate the passing of years as game character grows up
    #made up of print statement, declaration of object and while loop
    #while loop breaks as soon as years are counted up from 1 to 5
    ################################################################################################

    print (f'Years pass and you slowly begin to grow up.')
    
    years = 1
    
    while years <= 5:
        print(years)
        time.sleep(1)
        years += 1
    
    #print setting the scene for scene 3
    print(f"""
It is time for you to take back your kingdom. You go back to the African 
Planes to confront your uncle and his army of hyenas. To win the fight you have
to prepare to fight the right number of hyenas.""" )

    #delayed user prompt for better story flow
    time.sleep(3)

    #try and except <further down> statements to ensure code functionality despite wront input
    try:
    
        #prompt user to guess number of hyenas  
        hyenas = int((input(prompt = """
How many hyenas are you capable of fighting? 
<Answer in digits>
        """)))

        #nested conditional statement to evaluate answer
        #different print messages depending on the number guessed
        #if guess is under 50, user fails game
        if hyenas >= 50:
            if hyenas > 100:
                print(f""" 
{hyenas} hyenas is a bit ambitious, luckily there won't be quite as many to fight.
Good luck in battle!
                """)
                win()

            elif hyenas >= 50 or hyenas > 80:
                print(f"""
That is a good guess. There are more or less {hyenas} hyenas at the African Plane.
Good luck in battle!
                """)
                win()

        elif hyenas < 50:
            print(f"""
When you reach the African Plan you realize there are more than {hyenas} hyenas.
Despite your greatest efforts you cannot beat them. 
            """)
            fail()

        else:
            print('Invalid entry. Try again')
            scene_3()
            
    except:
        print('Invalid entry. Try again')
        scene_3()
                
#defining win function
def win():
    
    #print fight message
    print("""
___       ___     __       ___ ___       ___     __   __               ___       __   ___  __  
 |  |__| |__     |__)  /\   |   |  |    |__     /  ` /  \  |\/|  |\/| |__  |\ | /  ` |__  /__` 
 |  |  | |___    |__) /~~\  |   |  |___ |___    \__, \__/  |  |  |  | |___ | \| \__, |___ .__/ 
                                                                                             """)
    
    #delayed user prompt for better story flow
    time.sleep(5)
    
    #print win message
    print(f"""
{('*^')*40}

Your hard work has paid off! Congratulations the African Plane is free from its evil 
reign. You have taken over the throne as rightful king. 
You have won.

{('*^')*40}
    """)
    
    #prompt user to either end game or start again
    end_choice = input(prompt = "Type 1 to exit or 2 to start from the beginning")
    
    if end_choice == "1" in end_choice:
        exit(0)
        
    elif end_choice =="2" in end_choice:
        scene_1()
        
#defining fail function
def fail():
    
    #print fail message
    print(f"""{('*^')*40}  
You have gone to lion heaven.

<Game Over>
    {('*^')*40} """)
    
    #prompt user to either end game or start again
    
    end_choice = input(prompt = "Type 1 to exit or 2 to start from the beginning")
    if end_choice == "1" in end_choice:
        exit(0)
    elif end_choice =="2" in end_choice:
        scene_1()
        
scene_1()
```
